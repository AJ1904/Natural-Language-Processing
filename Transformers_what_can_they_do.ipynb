The fundamental element within the Transformers library is the `pipeline()` function. This function links a model with its required preprocessing and postprocessing procedures, enabling direct input of text and receiving a coherent response.

The default setting for a pipeline involves selecting a specific pretrained model tailored for sentiment analysis in English, which is downloaded and stored in cache upon creating the classifier object. Subsequent runs of the command will utilize the cached model without requiring another download.



```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a HuggingFace course my whole life.")
```

```python
classifier("I've been waiting for death my whole life.")
```

Multiple sentences can be passed at once.

```python
classifier(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
)
```

When text is passed to a pipeline, three primary steps are involved:

1. Preprocessing the text into a format understandable by the model.
2. Passing the preprocessed inputs to the model.
3. Post-processing the model's predictions to comprehend the outcomes.

Some of the available pipelines include:

- `feature-extraction`
- `fill-mask`
- `ner` (named entity recognition)
- `question-answering`
- `sentiment-analysis`
- `summarization`
- `text-generation`
- `translation`
- `zero-shot-classification`


# zero-shot-classification

Sometimes, we need to figure out what category a text belongs to, but we don't have labels for it. The zero-shot-classification pipeline helps with this. It lets us pick our own labels for classification, instead of just using the ones the model already knows. This method doesn't need us to fine-tune the model on our data. It can give scores for different labels without any special training.



```python
classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

# text-generation
Using a pipeline to create text is like giving the model a starting point, and it continues writing from there. It's a bit like the predictive text feature on phones. Generating text involves some randomness, so you might not always get the exact same results. You can adjust how many versions of the text are created using `num_return_sequences` and control the length of the output using `max_length`.


```python
generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

A particular model can also be selected instead of using default.

```python
generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

# fill-mask
The job of fill-mask pipeline is to fill in the blanks in a provided text.
- When using this, the `top_k` argument helps in determining how many options you'd like to see displayed.
- It is important to note that the model fills in a special word, usually denoted as `<mask>`, which is commonly referred to as a mask token.
- Different models might use different tokens for masking, so it is a good practice to confirm the correct mask token when exploring various models.
- We can typically identify the mask word by checking the widget or interface.


```python
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

# ner
Named Entity Recognition (NER) involves the model identifying entities like people, places, or organizations within a text.



```python
ner = pipeline("ner", grouped_entities=True)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

- The model correctly identified Sylvain as a person (PER), recognized Hugging Face as an organization (ORG), and identified Brooklyn as a location (LOC).
- To ensure entities are grouped together, we use the option `grouped_entities=True` when creating the pipeline. This instructs the pipeline to consolidate parts of the sentence that belong to the same entity.
- In the given example, despite "Hugging Face" being a multi-word entity, the model correctly grouped it as a single organization.
- Notably, during preprocessing, certain words might be broken into smaller parts, such as Sylvain being split into four segments: S, ##yl, ##va, and ##in. However, the pipeline effectively reassembles these fragments during post-processing.

# question-answering
The question-answering pipeline functions by deriving answers from the provided context rather than generating them.


```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```

# summarization
Summarization involves condensing a text into a shorter version while maintaining the crucial elements from the original. Similar to text generation, we can set parameters like `max_length` or `min_length` to control the length of the resulting summary.


```python
summarizer = pipeline("summarization")
summarizer(
    """
    America has changed dramatically during recent years. Not only has the number of
    graduates in traditional engineering disciplines such as mechanical, civil,
    electrical, chemical, and aeronautical engineering declined, but in most of
    the premier American universities engineering curricula now concentrate on
    and encourage largely the study of engineering science. As a result, there
    are declining offerings in engineering subjects dealing with infrastructure,
    the environment, and related issues, and greater concentration on high
    technology subjects, largely supporting increasingly complex scientific
    developments. While the latter is important, it should not be at the expense
    of more traditional engineering.

    Rapidly developing economies such as China and India, as well as other
    industrial countries in Europe and Asia, continue to encourage and advance
    the teaching of engineering. Both China and India, respectively, graduate
    six and eight times as many traditional engineers as does the United States.
    Other industrial countries at minimum maintain their output, while America
    suffers an increasingly serious decline in the number of engineering graduates
    and a lack of well-educated engineers.
"""
)
```

# translation
For translation, we can either use a default model by specifying a language pair in the task name (e.g., `translation_en_to_fr`), or simply choose a model from the Model Hub. We can define parameters like `max_length` or `min_length` to control the length of the translated output.


```python
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
translator("Ce cours est produit par Hugging Face.")
```

# Bias and Limitations

While pretrained or fine-tuned models are potent tools, they do have limitations. One significant aspect is that for extensive pretraining, researchers often scrape diverse content from the internet, encompassing both high-quality and problematic data.

For instance, when prompted to fill in missing words in sentences, a model might offer only one gender-neutral response ("waiter/waitress"), while other options might reflect work occupations often associated with a specific gender. Surprisingly, even in models like BERT, trained on apparently neutral datasets (English Wikipedia and BookCorpus), there can be associations leading to biased outputs.

It's crucial to bear in mind that when utilizing these tools, the original model might generate content embodying biases like sexism, racism, or homophobia. Fine-tuning the model on your data won't inherently eliminate these biases.


```python
unmasker = pipeline("fill-mask", model="bert-base-uncased")
result = unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])

result = unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])
```
